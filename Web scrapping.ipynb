```python
soup.find_all('div')
country_all=soup.find_all('div', class_="col-md-4 country")
```

```python
country_all[:2]
```

```python
with open('countries_info.csv', 'w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['Country', 'Capital', 'Population', 'Area'])

    for country_1 in country_all:
        country_name = country_1.find('h3', class_='country-name').text.strip()
        capital_name = country_1.find('span', class_='country-capital').text.strip()
        population = country_1.find('span', class_='country-population').text.strip()
        area = country_1.find('span', class_='country-area').text.strip()

        writer.writerow([country_name, capital_name, population, area])

        print(country_name, capital_name, population, area)
```
